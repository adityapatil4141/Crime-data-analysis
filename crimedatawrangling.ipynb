# ** Q1. How many violent crimes have happened from year 1960 - 2000**

---



```python

#Checking for any duplicate values in the column:

df.sort_values("Violent crime total", inplace = True)

bool_series = df["Violent crime total"].duplicated()

df[bool_series]
```

```python

# Taking necessary columns and assingning variable to it:

Violent_crimes_df = df[['Year','Violent crime total']]
```

```python

#Printing newly made dataframe:

Violent_crimes_df.head()
```

```python

# sorting data according to required year span:

Violent_crimes_1960_2000 =  Violent_crimes_df [Violent_crimes_df ['Year']<=2000]
```

```python

# printing the output:

Violent_crimes_1960_2000.sort_values(by="Year")
```

Q

```python

#Total crimes happened from year 1960-2000:

Violent_crimes_1960_2000['Violent crime total'].sum()
```

```python
#Visualization:

#Bar visualization for violent crimes in time span of 1960 - 2000

Violent_crimes_1960_2000.plot('Year', 'Violent crime total', kind = 'bar',color = ('r'), edgecolor = 'black')
plt.title("Year X violent crime")
plt.xlabel("Years")
plt.ylabel("Violent crime total")

fig = plt.figure()
plt.rcParams['figure.figsize'] = [15,5]

plt.show()
```

```python

# Plotting data in another form (scatter plot) :

sns.scatterplot(x = Violent_crimes_1960_2000['Year'],y= Violent_crimes_1960_2000['Violent crime total'])
```
